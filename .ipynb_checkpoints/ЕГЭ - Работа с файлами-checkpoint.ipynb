**Тип 10 № 76113**

```python
f = open("local/Гончаров.txt", 'r', encoding='utf-8')
l = f.read()
zp = '.,!?«»:;()"[]{}…'
for i in zp:
   l=l.replace(i, '')
l =l.replace('\n', ' ')
l =l.replace('-', ' ')
l = l.split()

c = 0
for i in l:
    a = i.lower()
    lena = len(i)
    if (a[0] == 'м' or a[0] == 'm') and (a[-1] == 'а' or a[-1] == 'a') and (lena>= 4):
        c +=1
        # print(a)
print(c)

f.close()
```

Тип 17 № 
75281

```python
f=open('local/17.txt', encoding = 'utf-8')
h = f.read()

contra = 0
h = h.split('\n') 
h =[int(i) for i in h if len(i)>0]
maxim = max(h)
minim = min(h)
summ = 0
for i in range(len(h)-2):
    a = h[i]
    b = h[i+1]
    c = h[i+2]
    if (10>a/1000>=1) or (10>b/1000>=1) or (10>c/1000>=1):
        if ((a%5 == maxim%5) +(b%5 == maxim%5) +(c%5 == maxim%5))<=1:
            if  ((a%7 == minim%7) +(b%7 == minim%7) +(c%7 == minim%7))>=2:
                contra += 1
                if (a+b+c)>summ:
                    summ = a+b+c

#print(h)
print(contra, summ)
f.close()
```
